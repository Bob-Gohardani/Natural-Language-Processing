```python
# Counter class
from collections import Counter

# Counter class can keep track of all given languages
c = Counter(['python', 'java'])
print(c)
c.update(['c++', 'python'])
print(c)
c.update(['javascript', 'python'])
print(c)
```

```python
with open('./Files/data.csv') as csv_file:
    # read the csv file as a dictionary
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

print(language_counter)
print("")
print(language_counter.most_common(15))

# another way
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

print(languages, popularity)
```

```python
# reverse the other of both X, Y values, it modifies the list and doesnt return a new list
languages.reverse()
popularity.reverse()

# plt.barh() creates a horizontal bar chart
plt.barh(languages, popularity)

plt.xlabel('number of users')
plt.title('Most popular languages')

plt.tight_layout()
plt.show()
```

# Filling Area On Line Plots

```python
data = pd.read_csv("./Files/data_1.csv")
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']

plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label='All Devs')
plt.plot(ages, py_salaries, label="Python")

overall_median = 57287
# fill the space between the py_salaries line plot and the median salary
plt.fill_between(ages, 
                 py_salaries,
                 overall_median, 
                 alpha=0.5,
                 where=(py_salaries > overall_median), # only fill values bigger than the overall_median
                 interpolate=True, # this is for the plot to show correctly with where condition
                 label='Above AVG')

plt.fill_between(ages, 
                 py_salaries, 
                 overall_median, 
                 alpha=0.5,
                 where=(py_salaries <= overall_median),
                 interpolate=True, 
                 label='Below AVG')

plt.legend()
plt.title("Median salary (USD) by Age")
plt.xlabel("Age")
plt.ylabel("Salary (USD)")

plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv("./Files/data_1.csv")
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']

plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label='All Devs')
plt.plot(ages, py_salaries, label="Python")

plt.fill_between(ages, 
                 py_salaries, 
                 dev_salaries,
                 alpha=0.5,
                 where=(py_salaries > dev_salaries), 
                 interpolate=True,)

plt.fill_between(ages, 
                 py_salaries, 
                 dev_salaries,
                 alpha=0.5,
                 where=(py_salaries <= dev_salaries),
                 interpolate=True)

plt.legend()
plt.title("Median salary (USD) by Age")
plt.xlabel("Age")
plt.ylabel("Salary (USD)")

plt.tight_layout()
plt.show()
```

# Time Series Plotting
    plotting dates

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates, y, linestyle="solid")

# we get the figure from the plot then formate the date so it looks kinda vertical
plt.gcf().autofmt_xdate()
# make the date to the format of Month (name) , day , year
date_format = mpl_dates.DateFormatter('%b, %d %Y')
# get the axises, then x axis, then set the fomat of x axis same as date_format
plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('./Files/bitcoin.csv')
# since the dates that we get are strings, we transform them into datetime format
data['Date'] = pd.to_datetime(data['Date'])
# then we sort them based on the date, so there is no error in dates order
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle="solid")

plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Prices')

plt.tight_layout()
plt.show()
```

# Live Data Plotting

```python
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# create two empty lists
x_vals = []
y_vals = []

# use count class, each time you call next() it will increase the integer by one
index = count()

def animate(i):
    # add to x, y lists
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    
    # clear the plot from last time (cla: clear axis)
    plt.cla()
    plt.plot(x_vals, y_vals)

# FuncAnimation() class allows live plotting, it haa arguments, current plot, plotting function, 
# interval : how many times per milliseconds update the plot
ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()
plt.show()
```

# Subplots

<br>subplots help when you have several plots to show at once<br>
    Figure : is the window (canvas) that shows all plots<br>
    axis : the actual plots

```python
# create one figure with two axes, both in same row
# sharex=True  : only show info of X-axis in lower plot
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages_x, dev_y, label="All devs")

ax2.plot(ages_x, py_dev_y, label="Python")
ax2.plot(ages_x, js_dev_y, label="JS")


ax1.legend()
ax1.set_ylabel('Salary (USD)')
ax1.set_title('Median salary (USD) by Age')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
# create two different figures where each of them has one axis
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages_x, dev_y, label="All devs")

ax2.plot(ages_x, py_dev_y, label="Python")
ax2.plot(ages_x, js_dev_y, label="JS")


ax1.legend()
ax2.set_xlabel('Ages')
ax1.set_ylabel('Salary (USD)')
ax1.set_title('Median salary (USD) by Age')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Salary (USD)')
ax1.set_title('Median salary (USD) by Age')

plt.tight_layout()
plt.show()

fig1.savefig('fig1.png')
fig2.savefig('fig2.png')
```
